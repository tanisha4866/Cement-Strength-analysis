# **outliers capping**

```python
outliers = ['Blast Furnace Slag ','Water  ','Superplasticizer ','Fine Aggregate ','Age ']
```

```python
def outlier_capping(dataframe :pd.DataFrame,outliers:list):
  df = dataframe.copy()
  for i in outliers:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iq3 = q3 -q1
    upper_limit=q3 + 1.5*iq3
    lower_limit = q3 - 1.5*iq3
    df.loc[df[i]>upper_limit,i]=upper_limit
    df.loc[df[i]<lower_limit,i]=lower_limit
  return df

df= outlier_capping(dataframe= df,outliers=outliers)
```

```python
plt.figure(figsize = (15,15),facecolor='white')
plotnumber = 1
for i in df.columns:
  ax = plt.subplot(4,3,plotnumber)
  sns.boxplot(df[i])
  plt.xlabel(i,fontsize=10)
  plotnumber +=1
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.pipeline import make_pipeline
```

```python
def check_model_performance(preprocessor,xtrain,xtest,ytrain,ytest):
  models = {'LinearRegression': LinearRegression(),
            'Ridge Regression': Ridge(alpha=1),
            'Lasso Regression': Lasso(alpha =1),
            'Random  Forest Regression': RandomForestRegressor(max_depth = 5),
            'Gradient Boosting regression': GradientBoostingRegressor(learning_rate=0.1)}
  for model_name,model in models.items():
    pipeline = make_pipeline(preprocessor,model)
    pipeline.fit(xtrain,ytrain)
    y_pred= pipeline.predict(xtest)
    mse =  mean_squared_error(ytest,y_pred)
    r2 = r2_score(ytest,y_pred)
    print(f'{model_name} - mse = {mse:.2f}\n {model_name}- r2 = {r2:.2f}')
```

```python
preprocessor_01 = make_pipeline(KNNImputer(n_neighbors=3),StandardScaler ())
preprocessor_02 = make_pipeline(KNNImputer(n_neighbors=3),MinMaxScaler ())
preprocessor_03 = make_pipeline(KNNImputer(n_neighbors=3),RobustScaler ())
```

```python
check_model_performance(preprocessor_01,xtrain,xtest,ytrain,ytest)
print( f'*' * 10)
check_model_performance(preprocessor_02,xtrain,xtest,ytrain,ytest)
print( f'*' * 10)
check_model_performance(preprocessor_03,xtrain,xtest,ytrain,ytest)
```

```python
import warnings
warnings.filterwarnings("ignore")


from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [100,200],
              'learning_rate':[0.1,0.01],
              'max_depth': [5,3,7],
              'min_samples_split':[2,4],
              'min_samples_leaf':[1,2,3]}

gd_rg=GradientBoostingRegressor()

#scoring is necessory in case of regression problem
grid=GridSearchCV(gd_rg,param_grid,scoring='neg_mean_squared_error',cv=5,verbose=1 )
grid.fit(xtrain,ytrain)
```
